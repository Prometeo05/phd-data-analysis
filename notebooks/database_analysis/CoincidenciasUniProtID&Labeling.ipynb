```python
Este codigo permite comparar dos archivos FASTA y encontrar coincidencias.
 Generando un archivo de texto con las coincidencias encontradas IDs y Etiquetas.
```

```python
# Instalar Biopython en Google Colab (si no está instalado)
!pip install biopython

from Bio import SeqIO

# Definir función para comparar archivos FASTA
def comparar_fasta(fasta_uniprot, fasta_personalizado, salida_txt):
    uniprot_dict = {str(seq.seq): seq.id for seq in SeqIO.parse(fasta_uniprot, "fasta")}
    personalizado_dict = {str(seq.seq): seq.id for seq in SeqIO.parse(fasta_personalizado, "fasta")}

    with open(salida_txt, "w") as salida:
        salida.write("Identificador_UniProt\tIdentificador_Personalizado\tSecuencia\n")
        for secuencia, id_uniprot in uniprot_dict.items():
            if secuencia in personalizado_dict:
                id_personalizado = personalizado_dict[secuencia]
                salida.write(f"{id_uniprot}\t{id_personalizado}\t{secuencia}\n")

# Rutas de archivos en Colab (ajusta según sea necesario)
fasta_uniprot = "/content/DS_BAETH_CompleteDataset_OnlyIDs[716].FASTA"
fasta_personalizado = "/content/DS_BAETH_CompleteDataset_Labeled[716].fasta"
salida_txt = "/content/coincidencias.txt"

# Ejecutar comparación
comparar_fasta(fasta_uniprot, fasta_personalizado, salida_txt)

# Mostrar enlace para descargar el archivo de resultados
from google.colab import files
files.download(salida_txt)
```

```python
from Bio import SeqIO

def comparar_fasta(fasta_uniprot, fasta_personalizado, salida_fasta):
    # Cargar las secuencias de ambos archivos en diccionarios
    uniprot_dict = {str(seq.seq): seq.id for seq in SeqIO.parse(fasta_uniprot, "fasta")}
    personalizado_dict = {str(seq.seq): seq.id for seq in SeqIO.parse(fasta_personalizado, "fasta")}

    # Guardar coincidencias en formato FASTA
    with open(salida_fasta, "w") as salida:
        for secuencia, id_uniprot in uniprot_dict.items():
            if secuencia in personalizado_dict:
                id_personalizado = personalizado_dict[secuencia]
                salida.write(f">{id_uniprot}|{id_personalizado}\n{secuencia}\n")

# Rutas de archivos en Colab (ajusta según la ubicación)
fasta_uniprot = "/content/DS_BAETH_CompleteDataset_OnlyIDs[716].FASTA"
fasta_personalizado = "/content/DS_BAETH_CompleteDataset_Labeled[716].fasta"
salida_fasta = "/content/coincidencias.fasta"

# Ejecutar comparación
comparar_fasta(fasta_uniprot, fasta_personalizado, salida_fasta)

# Mostrar enlace para descargar el archivo de resultados
from google.colab import files
files.download(salida_fasta)
```

```python
# Instalar Biopython (si es necesario en Colab)
!pip install biopython

from Bio import SeqIO
from collections import defaultdict

def encontrar_secuencias_identicas(fasta_path, salida_txt):
    # Diccionario para almacenar secuencias normalizadas y sus identificadores
    secuencias_dict = defaultdict(list)

    # Leer el archivo FASTA y almacenar las secuencias procesadas
    for secuencia in SeqIO.parse(fasta_path, "fasta"):
        secuencia_normalizada = str(secuencia.seq).strip().upper()  # Eliminamos espacios y convertimos a mayúsculas
        secuencias_dict[secuencia_normalizada].append(secuencia.id)

    # Guardar las coincidencias encontradas en un archivo de salida
    with open(salida_txt, "w") as salida:
        salida.write("Secuencia\tIdentificadores\n")
        for secuencia, ids in secuencias_dict.items():
            if len(ids) > 1:  # Solo guardar secuencias idénticas
                salida.write(f"{secuencia}\t{', '.join(ids)}\n")

# Ruta del archivo FASTA en Colab
fasta_path = "/content/coincidencias.fasta"  # Ajusta según la ubicación del archivo
salida_txt = "/content/secuencias_identicas.txt"

# Ejecutar la función
encontrar_secuencias_identicas(fasta_path, salida_txt)

# Descargar el archivo de resultados
from google.colab import files
files.download(salida_txt)
```
