Este codigo procesa un archivo fasta que contenga identificadores de UniProt tipo >P00687, extrae la informacion que existe en esa base de datos y crea un archivo de texto .txt con la coleccion de informacion y la secuencia.

```python
!pip install biopython
import requests
from Bio import SeqIO

def obtener_info_uniprot(identificador):
    url = f"https://rest.uniprot.org/uniprotkb/{identificador}.txt"
    respuesta = requests.get(url)
    if respuesta.status_code == 200:
        return respuesta.text
    else:
        return f"No encontrada para {identificador}"

# Procesa el archivo FASTA y busca información en UniProt
def procesar_fasta(fasta_path, salida_path):
    with open(salida_path, "w") as salida:
        for secuencia in SeqIO.parse(fasta_path, "fasta"):
            identificador = secuencia.id  # Se usa directamente el identificador
            info = obtener_info_uniprot(identificador)
            salida.write(f">{identificador}\n{info}\n\n")

# Ejecutar la función con los archivos deseados
procesar_fasta("DS_BAETH_CompleteDataset_OnlyIDs[716].FASTA", "resultados.txt")
```

Este codigo busca secuencias de proteinas en formato .fasta y devuelve un archivo de texto .txt que te indica cuales son.

```python
# Instalar Biopython (si es necesario en Colab)
!pip install biopython

from Bio import SeqIO
from collections import defaultdict

def encontrar_secuencias_identicas(fasta_path, salida_txt):
    # Diccionario para almacenar secuencias y sus identificadores
    secuencias_dict = defaultdict(list)

    # Leer el archivo FASTA y almacenar las secuencias
    for secuencia in SeqIO.parse(fasta_path, "fasta"):
        secuencias_dict[str(secuencia.seq)].append(secuencia.id)

    # Guardar las coincidencias encontradas en un archivo de salida
    with open(salida_txt, "w") as salida:
        salida.write("Secuencia\tIdentificadores\n")
        for secuencia, ids in secuencias_dict.items():
            if len(ids) > 1:  # Solo guardar secuencias idénticas
                salida.write(f"{secuencia}\t{', '.join(ids)}\n")

# Ruta del archivo FASTA en Colab
fasta_path = "/content/coincidencias.fasta"  # Ajusta según la ubicación del archivo
salida_txt = "/content/secuencias_identicas.txt"

# Ejecutar la función
encontrar_secuencias_identicas(fasta_path, salida_txt)

# Descargar el archivo de resultados
from google.colab import files
files.download(salida_txt)
```
